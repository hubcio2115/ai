```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
cities = pd.read_csv("./miasta.csv")
print(cities)
print()
print(cities.values)
```

```python
# type: ignore
cities.loc[10] = [2010, 460, 555, 405] 
print(cities)
```

```python
years = cities.Rok
gd = cities.Gdansk

plt.plot(years, gd, color="red", marker="o")
plt.legend("Ludność Gdańska")
plt.xlabel("Lata")
plt.ylabel("Ludność")
```

```python
pzn = cities.Poznan
szcz = cities.Szczecin

plt.plot(years, gd)
plt.plot(years, pzn)
plt.plot(years, szcz)

plt.xlabel("Lata")
plt.ylabel("Ludność")
```

```python
import numpy as np

def standardize(df):
  res = 0
  for val in df.values:
    res += (val - np.mean(df.values)) / np.std(df.values)

  return res

z_gdansk = standardize(cities.Gdansk)
z_poznan = standardize(cities.Poznan)
z_szczecin = standardize(cities.Szczecin)
```
