```python
import numpy as np
import pygad
import time
```

```python
items = [
    {
        "value": 100,
        "weight": 7
    },
    {
        "value": 300,
        "weight": 7
    },
    {
        "value": 200,
        "weight": 6
    },
    {
        "value": 40,
        "weight": 2
    },
    {
        "value": 500,
        "weight": 5
    },
    {
        "value": 70,
        "weight": 6
    },
    {
        "value": 100,
        "weight": 1},
    {
        "value": 250,
        "weight": 3
    },
    {
        "value": 300,
        "weight": 10
    },
    {
        "value": 280,
        "weight": 3
    },
    {
        "value": 300,
        "weight": 15
    }
]

gene_space = [0, 1]

def fitness_func(solution, solution_idx):
    values = []
    weights = []

    for i in range(len(items)):
        if solution[i] == 1:
            values.append(items[i]['value'])
            weights.append(items[i]['weight'])

    if np.sum(weights) > 25:
        return 0

    return np.sum(values)

sol_per_pop = 10
num_genes = len(items)

num_parents_mating = 5
num_generations = 30
keep_parents = 2

parent_selection_type = "rank"

crossover_type = "single_point"

mutation_type = "random"

mutation_percent_genes = 10
```

```python
# without stop

results = []
for _ in range(10):
    start = time.time()
    ga_instance = pygad.GA(
        gene_space=gene_space,
        num_generations=num_generations,
        num_parents_mating=num_parents_mating,
        fitness_func=fitness_func,
        sol_per_pop=sol_per_pop,
        num_genes=num_genes,
        parent_selection_type=parent_selection_type,
        keep_parents=keep_parents,
        crossover_type=crossover_type,
        mutation_type=mutation_type,
        mutation_percent_genes=mutation_percent_genes,
    )
    ga_instance.run()
    end = time.time()
    results.append(end-start)

mean1 = np.mean(results)
```

```python
# without stop

results = []
for _ in range(10):
    start = time.time()
    ga_instance = pygad.GA(
        gene_space=gene_space,
        num_generations=num_generations,
        num_parents_mating=num_parents_mating,
        fitness_func=fitness_func,
        sol_per_pop=sol_per_pop,
        num_genes=num_genes,
        parent_selection_type=parent_selection_type,
        keep_parents=keep_parents,
        crossover_type=crossover_type,
        mutation_type=mutation_type,
        mutation_percent_genes=mutation_percent_genes,
        stop_criteria=["reach_1600"]
    )
    ga_instance.run()
    end = time.time()
    results.append(end-start)

mean2 = np.mean(results)
```

```python
print(f"without stop: {mean1}s")
print(f"with stop: {mean2}s")
```
