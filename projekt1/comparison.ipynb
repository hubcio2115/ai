# Comparison

With both tests completed we can compare how do those two algorithms stand against each other.


```python
import pandas as pd
import numpy as np


df_GA = pd.read_csv('GA.csv', sep=';')
df_PSA = pd.read_csv('PSA.csv', sep=';')
```

```python
times_5x5_mean_GA = df_GA['5x5.times'].mean()
times_5x5_mean_PSA = df_PSA['5x5.times'].mean()

times_10x10_mean_GA = df_GA['10x10.times'].mean()
times_10x10_mean_PSA = df_PSA['10x10.times'].mean()

times_15x15_mean_GA = df_GA['15x15.times'].mean()
times_15x15_mean_PSA = df_PSA['15x15.times'].mean()

results_10x10_mean_GA = df_GA['10x10.results'].mean()
results_10x10_mean_PSA = df_PSA['10x10.results'].mean()

results_15x15_mean_GA = df_GA['15x15.results'].mean()
results_15x15_mean_PSA = df_PSA['15x15.results'].mean()
```

```python
from matplotlib import pyplot as plt

times_mean = [times_5x5_mean_GA, times_5x5_mean_PSA, times_10x10_mean_GA,
              times_10x10_mean_PSA, times_15x15_mean_GA, times_15x15_mean_PSA]
labels = ['5x5 GA', '5x5 PSA', '10x10 GA',
          '10x10 PSA', '15x15 GA', '15x15 PSA']
x = np.arange(len(labels))
width = 0.35

fig, ax = plt.subplots()
bars = ax.bar(x - width/2, times_mean, width)

ax.set_title('Times (ms)')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

bars[0].set_color('r')
bars[1].set_color('r')
bars[2].set_color('g')
bars[3].set_color('g')
bars[4].set_color('b')
bars[5].set_color('b')

fig.text(0.5, 0, '(the lower the better)',
         ha='center', va='center', fontsize=10)

fig.tight_layout()

plt.show()
```

From this plot we can surely say that PSA algorithm for this problem is much much slower. There is nothing to add to that.


```python
results_mean = [results_10x10_mean_GA, results_10x10_mean_PSA,
                results_15x15_mean_GA, results_15x15_mean_PSA]
labels = ['10x10 GA', '10x10 PSA', '15x15 GA', '15x15 PSA']
x = np.arange(len(labels))
width = 0.35

fig, ax = plt.subplots()
bars = ax.bar(x - width/2, results_mean, width)

ax.set_title('Results')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

bars[0].set_color('g')
bars[1].set_color('g')
bars[2].set_color('b')
bars[3].set_color('b')

fig.text(0.5, 0, '(the lower the better)',
         ha='center', va='center', fontsize=10)

fig.tight_layout()

plt.show()
```

But here we can see something interesting. Comparably we can see that PSA on average got much better results in case of fitness rating than GA got. We didn't include the first level of complexity since PSA got 0% of accuracy.

# Conclusion

The GA algorithm is a better way to solve this kind of problem for a low complexity problem, and PSA should be considered the better option for more complex ones.

In case of time GA got way ahead of PSA but in this kind of scenario where we want a better result than time of when it is given to us I'd say that the time of PSA is still acceptable.
